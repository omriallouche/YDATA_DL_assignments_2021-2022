```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import sklearn
%matplotlib inline
from PIL import Image
```

```python
im = np.asarray(Image.open('tornado.jpeg').resize((600,400)))/255.0
```

```python
im = 0.299*im[:,:,0] + 0.587*im[:,:,1] + 0.114*im[:,:,2]
```

```python
plt.imshow(im,cmap='gray'); plt.show()
```

```python
np.min(im)
```

```python
np.max(im)
```

```python
strech_im = im-np.min(im)
strech_im = strech_im/np.max(strech_im)
```

```python
np.min(strech_im)
```

```python
np.max(strech_im)
```

```python
plt.subplot(1,2,1); plt.imshow(im,cmap='gray'); plt.subplot(1,2,2); plt.imshow(strech_im,cmap='gray');plt.show()
```

```python
mean = np.mean(im)
std = np.std(im)
```

```python
im_normalize = (im - mean)/std
```

```python
im_normalize_new = im_normalize*std + mean
```

```python
im_normalize_new
```

```python
np.sqrt(np.sum(np.power(im-im_normalize_new,2))) 
```
